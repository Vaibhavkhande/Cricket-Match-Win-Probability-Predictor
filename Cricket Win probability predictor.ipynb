## 1.extracting current score from the deliveries dataset

```python
deliveries['current_score'] = deliveries.groupby(['match_id','inning']).cumsum()['total_runs']
```

## 2.extracting runs left from the deliveries dataset

```python
deliveries['runs_left'] = deliveries['target']-deliveries['current_score']
```

## 3.extrcting balls left from deliveries dataset

```python
deliveries['balls_left']= 126-(deliveries['over']*6+deliveries['ball'])
```

```python
deliveries.head()
```

## 4. extracting wickets left from deliveries dataset

```python
deliveries['player_dismissed'] = deliveries['player_dismissed'].fillna("0")
deliveries['player_dismissed'] = deliveries['player_dismissed'].apply(lambda x:x if x == "0" else "1")
deliveries['player_dismissed'] = deliveries['player_dismissed'].astype('int')
wickets = deliveries.groupby('match_id').cumsum()['player_dismissed'].values
deliveries['wickets'] = 10 - wickets
deliveries.head()
```

## 5.extrcting current runret from deliveries dataset

```python
## crr = run/over
deliveries['crr'] =(deliveries['current_score']*6)/(120 - deliveries['balls_left'])
```

## 6.extracting required runret from deliveries dataset

```python
#rrr =runs_left/balls_left
deliveries['rrr'] = (deliveries['runs_left']*6)/deliveries['balls_left']
```

```python
deliveries.head()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
deliveries['result'] = deliveries.apply(result,axis=1)
```

```python
final_data = deliveries[['batting_team','bowling_team','city','runs_left','balls_left','wickets','target','crr','rrr','result']]
```

```python
final_data.head()
```

```python
final_data = final_data.sample(final_data.shape[0])
```

```python
final_data.head()
```

```python
final_data['batting_team']=final_data['batting_team'].str.replace('Delhi Daredevils','Delhi capitals')
final_data['bowling_team']=final_data['bowling_team'].str.replace('Delhi Daredevils','Delhi capitals')

final_data['batting_team']=final_data['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
final_data['bowling_team']=final_data['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
final_data = final_data[final_data['batting_team'].isin(teams)]
final_data = final_data[final_data['bowling_team'].isin(teams)]
```

```python
final_data.isnull().sum()
```

```python
final_data.dropna(inplace=True)
```

```python
final_data['batting_team'].unique()
```

```python
final_data=final_data[final_data['balls_left'] !=0]
```

```python
x = final_data.iloc[:,:-1]
y = final_data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder = 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
     ('step1',trf),
     ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
